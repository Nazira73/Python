```python
from sklearn.datasets import load_iris
import pandas as pd
%matplotlib inline
iris_dataset = load_iris()
```

```python
iris_dataset
```

```python
print("Keys of iris_dataset: \n{}".format(iris_dataset.keys()))
```

```python
print(iris_dataset['DESCR'] + "\n...end")
```

```python
print("Target names: {}".format(iris_dataset['target_names']))
```

```python
print("Shape of data: {}".format(iris_dataset['data'].shape))
iris_dataset['data']
```

```python
df_pred = pd.DataFrame(iris_dataset['data'],columns=iris_dataset['feature_names'])
```

```python
df_pred.head()
```

```python
df_response = pd.DataFrame(iris_dataset['target'].astype('object'),columns = ['class'])
df_response.head()
```

```python
df = pd.concat([df_pred,df_response],axis = 1)
print("Shape of df: {}".format(df.shape))
```

```python
from sklearn.model_selection import train_test_split
df_train,df_test = train_test_split(df,random_state = 0)
```

```python
x_train = df_train.drop(['class'],axis = 1)
y_train = df_train['class']

x_test = df_test.drop(['class'],axis = 1)
y_test = df_test['class']
```

```python
print("Shape of train predictors: {}".format(x_train.shape))
print("Shape of test predictors: {}".format(x_test.shape))
print("Shape of train response: {}".format(y_train.shape))
print("Shape of test response: {}".format(y_test.shape))
```

```python
y_train = y_train.astype(int)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(x_train,y_train)
```

```python
pred = knn.predict(x_test)
```

```python
pd.crosstab(pred,y_test)
```

```python
y_test = y_test.astype(int)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,pred)
```
